```python
#Importações das Bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```


#Notebook de Clustering com K-Means no Conjunto de Dados Iris

Neste notebook, iremos explorar o algoritmo de clustering K-Means aplicado ao conjunto de dados Iris. O conjunto de dados Iris contém informações sobre três espécies de flores de íris, medindo as dimensões das sépalas e pétalas.

#O que é o algoritmo de clustering K-Means e para que ele serve?


O algoritmo de clustering K-Means é uma técnica popular de aprendizado não supervisionado utilizada para agrupar pontos de dados sem rótulos em clusters distintos. Ele pertence à categoria de algoritmos particionais, onde os dados são divididos em subconjuntos, conhecidos como clusters, com o objetivo de minimizar a variância intra-cluster e maximizar a variância inter-cluster.


O algoritmo de clustering K-Means é utilizado para agrupar dados sem rótulos em clusters, com o objetivo de identificar padrões intrínsecos nos dados. Ele serve para várias finalidades em ciência de dados e análise exploratória de dados.

**Ex: Em uma base de dados comercial, que é muito grande, é normal que você tenha vários clusters. Pois assim a visualização e retorno dos dados é mais eficiente.**

A clusterização é uma técnica muito usada em projetos de Data Science e Machine Learning para segmentação de clientes, divisão de produtos, análises de redes sociais, visualização de dados e outros fins. A pessoa cientista de dados precisa conhecer essas soluções, pois elas são extremamente importantes para diversas necessidades do dia a dia.

```python
iris = load_iris()
data = iris.data

plt.scatter(data[:, 0], data[:, 1], s=50)
plt.title('Dados para Clustering (Iris)')
plt.xlabel('Comprimento da Sépala (cm)')
plt.ylabel('Largura da Sépala (cm)')
plt.show()

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)


```

O conjunto de dados Iris é carregado, contendo informações sobre as dimensões da sépala e pétala de três espécies de flores de íris. Em seguida, uma visualização gráfica é criada utilizando um gráfico de dispersão, onde cada ponto representa uma flor de íris com base no comprimento e largura da sépala.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Método Elbow para Escolha de K (Iris)')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Soma dos Quadrados das Distâncias')
plt.show()

```

A curva Elbow, que representa a relação entre o número de clusters e o WCSS, é plotada utilizando matplotlib. O ponto de "cotovelo" na curva sugere o número ótimo de clusters, facilitando a interpretação e a configuração adequada do modelo K-Means para análise de agrupamento no conjunto de dados Iris.
