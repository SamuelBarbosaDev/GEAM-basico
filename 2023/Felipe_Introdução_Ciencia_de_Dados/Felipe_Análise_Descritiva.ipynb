A análise univariada é útil para entender a natureza e as propriedades individuais de cada variável, identificar possíveis outliers ou valores extremos e obter insights iniciais sobre os dados.

#**Variáveis Categóricas**

Variáveis categóricas são tipos de variáveis que representam categorias ou rótulos discretos em vez de valores numéricos. Elas podem ser nominais (sem uma ordem específica) ou ordinais (com uma ordem implícita). Exemplos de variáveis categóricas incluem gênero, cor dos olhos, tipo de animal ou classificação por nível educacional. Para analisar variáveis categóricas, geralmente utilizamos tabelas de frequência, gráficos de barras ou gráficos de pizza para visualizar a distribuição das categorias.

```python
# Contagem de cada categoria em uma variável categórica
print(data['Species'].value_counts())

# Gráfico de barras para visualizar a distribuição das categorias
plt.bar(data['Species'].value_counts().index, data['Species'].value_counts().values)
plt.xlabel('Espécies')
plt.ylabel('Contagem')
plt.title('Gráfico de Barras da Quantidade de Espécies')
plt.show()
```

É realizada uma contagem desses dados categoricos, a fim de uma melhor visualização na hora de realizar estatísticas. Nesse caso trabalhando com as espécies de flores presentes no intuito de facilitar quem vai realizar cálculos com esses valores. Essas variáveis possuem uma ordem natural e podem ser submetidas a operações matemáticas como adição, subtração, multiplicação e divisão. Nas análises descritivas, é comum calcular estatísticas resumidas, como média, mediana, desvio padrão, entre outras, para resumir a distribuição dos dados numéricos.

#**Variáveis Bivariadas**

Variáveis bivariadas são aquelas que envolvem a análise de duas variáveis simultaneamente para explorar as relações entre elas. Em outras palavras, estamos interessados em como uma variável está relacionada ou influencia outra variável. A análise bivariada é essencial para identificar associações ou padrões entre duas variáveis e analisar como uma variável pode afetar a outra em um conjunto de dados.

```python
# Gráfico de dispersão para analisar a relação entre duas variáveis numéricas
plt.scatter(data['SepalLengthCm'], data['PetalLengthCm'])
plt.xlabel('Comprimento da Sépala (cm)')
plt.ylabel('Comprimento da Pétala (cm)')
plt.title('Gráfico de Dispersão')
plt.show()

# Calcular a média do comprimento da sépala por espécie
mean_sepal_length_by_species = data.groupby('Species')['SepalLengthCm'].mean()

# Gráfico de barras para a média do comprimento da sépala por espécie
plt.bar(mean_sepal_length_by_species.index, mean_sepal_length_by_species.values)
plt.xlabel('Espécies')
plt.ylabel('Comprimento Médio da Sépala (cm)')
plt.title('Gráfico de Barras da Média do Comprimento da Sépala por Espécie')
plt.show()

# Tabela de contingência para analisar a relação entre duas variáveis categóricas
contingency_table = pd.crosstab(data['Species'], data['PetalLengthCm'])
print(contingency_table)
```

Essas análises e visualizações são úteis para entender a relação entre as diferentes variáveis numéricas e categóricas em relação às espécies de flores presentes no conjunto de dados("Iris"). Nesse exemplo, é utilizado o comprimento das pétalas pelas sépalas, essa relação nos ajudará a entender como essas duas variáveis se comportam para diferentes espécies de flores, depois é calculada a média dessa quantidade e então criada uma tabela de contingência que irá finalizar demonstrando a quantidade.
