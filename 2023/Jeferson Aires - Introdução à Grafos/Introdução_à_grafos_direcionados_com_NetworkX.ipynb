### Grafos Direcionados

Em alguns problemas em grafos, as relacões a serem modeladas não são
simétricas. Por exemplo, como representar seguidores em redes sociais?

imagem_2023-08-19_213355349-removebg-preview.png

* A Ana segue Beto e o Eduardo.
* Ninguém segue a Ana.
* O Daniel é seguido pelo Carlos e pelo Felipe.
* O Eduardo segue o Beto que o segue de volta.



Um grafo direcionado ou digrafo $G$ é um par ordenado $(V, E)$ tal que:

* $V$ é um conjunto vértices finito e não vazio.
* $E$ é um conjunto de pares ordenados de vértices. Os elementos de $E$ são
chamados de arestas. Cada aresta $(u, v) ∈ E$ é tal que $u$ indica origem e
$v$ indica destino. A aresta sai de $u$ e entra em $v.$

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

### Criando um Objeto NetworkX

Como vimos anteriormente, um grafo é composto por vértices conectados por arestas. NetworkX permite que vértices possam ser strings de texto, números, entre outros. O código a seguir, cria um objeto networkx.

```python
# Grafo direcionado
graph = nx.DiGraph()
```

### Adicionando Vértices ao Grafo (não direcionado)

Com nosso grafo criado, vamos adicionar vértices e arestas, para conectá-lo.

```python
# adicionando os vértices
graph.add_node(1)
graph.add_node(2)
graph.add_node(3)
graph.add_node(4)
graph.add_node(5)
graph.add_node(6)

# exibindo o número de vértices
print(graph.nodes())
print("\n")

# adicionando arestas
graph.add_edge(2, 1)
graph.add_edge(2, 5)
graph.add_edge(1, 3)
graph.add_edge(1, 5)
graph.add_edge(5, 6)
graph.add_edge(6, 4)
graph.add_edge(6, 3)

# Podemos imprimir o grafo usando o comando draw.
nx.draw(graph, with_labels = True)
```

### Removendo Vértices do Grafo (não direcionado)

Quando estamos trabalhando com grafos não direcionado podemos remover um vértice usando a função remove_node( ):

---

```python
# exibindo o número de vértices
print(graph.nodes())
print("\n")

# removendo o nó 2
graph.remove_node(6)

# grafo atualizado
print(graph.nodes())
```

### Obtendo os vizinhos de um nó

> Conhecendo o conceito de vizinhança em grafos direcionados.




imagem_2023-08-19_213355349-removebg-preview.png

* Dado $G = (V, E)$ direcionado e $(u, v) ∈ E$, dizemos que $(u, v)$ sai do
vértice $u$ e entra ou incide no vértice $v$.
* Dizemos também que $v$ é adjacente a $u.$
* Dizemos que $u$ e $v$ são os extremos da aresta $(u, v).$
* Vértices adjacentes a um vértice $u$ formam a vizinhança de $u$ e são
chamados vizinhos de $u.$


*No exemplo acima, Felipe é vizinho de Eduardo e Carlos, mas Ana não possui vizinhos.*

Para obter os vizinhos de um nó em um grafo direcionado, podemos usar os métodos $successors()$ e $predecessors().$ Os sucessores de um nó são os nós para os quais há uma aresta saindo do nó em questão, enquanto os predecessores são os nós dos quais há uma aresta chegando ao nó em questão.



```python
# Obter os sucessores e predecessores de um nó
node = 1
successors = list(graph.successors(node))
predecessors = list(graph.predecessors(node))

print(f"Sucessores do nó {node}: {successors}")
print(f"Predecessores do nó {node}: {predecessors}")
```

### Obtendo o grau de um nó direcionado

Em um grafo direcionado, o conceito de grau de um nó é dividido em duas medidas distintas: grau de entrada e grau de saída. Isso ocorre porque, nas arestas direcionadas, há uma distinção entre as direções em que as arestas estão indo e saindo de um nó específico.

* Grau de Entrada - O grau de entrada de um nó em um grafo direcionado é o número de arestas que chegam ao nó.
* Grau de Saída - O grau de saída de um nó em um grafo direcionado é o número de arestas que saem do nó.

Ou seja, o grau de um nó em um grafo direcinado é dado pela a soma de:
<center>
  $Grau \ de \ Entrada + Grau \ de \ Saída$
</center>

Para obter o grau de um nó em um determinado grafo, você pode usar a função $degree()$ da seguinte forma:

```python
# determinando seu grau
degree = graph.degree(node)
print(f"Grau do nó {node}: {degree}")
```
