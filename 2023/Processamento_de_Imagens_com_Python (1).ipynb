### Processamento de Imagens com Python - Ciência de Dados na Prática

---
Neste notebook falaremos sobre alguns métodos para o processamento de imagens, utilizando a biblioteca Pillow, que é uma poderosa biblioteca de processamento de imagens em Python. Ela é uma bifurcação (fork) da biblioteca Python Imaging Library (PIL) e fornece uma ampla gama de funcionalidades para manipular imagens. Com o Pillow, podemos abrir, salvar, redimensionar, cortar, girar, aplicar filtros e realizar diversas transformações em imagens de maneira simples e eficiente.

### Convertendo Imagens para arrays Numpy

```python
from matplotlib import image
from matplotlib import pyplot
```

Para carregar uma imagem como um array NumPy

```python
data = image.imread("sample_data/images/sao-paulo.jpg")
print(data)
```

Imprimindo as propriedades do array de pixels

```python
# Imprimindo o tipo de dado
print('Tipo de dado:', data.dtype)
# Forma do array
print('Forma do array:', data.shape)
# Valor máximo de pixels
print('Valor máximo de pixels:', data.max())
# Valor mínimo de pixels
print('Valor mínimo de pixels:', data.min())
```

### Exibindo o array de pixels como uma imagem

```python
pyplot.imshow(data)
```

Convertendo array de pixels em um objeto Image Pillow

```python
image2 = Image.fromarray(data)
```

Verificando o tipo do objeto

```python
type(image2)
```

Convertendo um objeto do tipo imagem para um array NumPy

```python
from numpy import asarray
```

```python
image = Image.open("sample_data/images/sao-paulo.jpg")
```

Método asarray que realiza a conversão

```python
data = asarray(image)
print(data.dtype)
print(data.shape)
```

### Salvando Imagens em Disco

Salvando a imagem no formato PNG

```python
image.save("sample_data/images/sao-paulo.png", format="PNG")
```

Salvando a imagem no formato GIF

```python
image.save("sample_data/images/sao-paulo.gif", format="GIF")
```

### Convertendo Imagens

Convertendo imagens em escala cinza

```python
image_cinza = image.convert(mode="L")
image_cinza
```

### Resize de Imagens com Python

Gerando thumbnail de tamanho 100x100

```python
image.thumbnail((100,100))
print('Dimensões da Imagem:', image.size)
image

# Perceba que ele mantém a porcentagem de dimensão da imagem, para que a mesm não seja distorcida.
```

Carregando a imagem novamente

```python
image = Image.open("sample_data/images/sao-paulo.jpg")

image_resize = image.resize((200,200))
print('Dimensões da Imagem:', image_resize.size)
pyplot.imshow(image_resize)
```

### Invertendo Imagens (Flip)

Inversão horizontal

```python
horizontal_image = image.transpose(Image.FLIP_LEFT_RIGHT)
pyplot.imshow(horizontal_image)
```

Inversão Vertical

```python
vertical_image = image.transpose(Image.FLIP_TOP_BOTTOM)
pyplot.imshow(vertical_image)
```

### Cortar Imagens (Crop)

```python
pyplot.imshow(image)
```

Exibindo o corte a partir das coordenadas específicas - Left, Upper, Right and Lower

```python
pyplot.imshow(image.crop((200, 200, 300, 300)))
```
